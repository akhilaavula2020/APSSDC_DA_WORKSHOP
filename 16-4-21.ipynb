## Robust scaler
 -  if your dataset contain more no of outliers
## Formula
# xi-Q2(x)/Q3(x)-Q1(x)
 - inter Quartile range(IQR)
 - Q1 ---> 25% (lower quartile)
 - Q2 ---->50%(medium quartile)
 - Q3 ----> 75%(upperquartile)

```python
import pandas as pd
import numpy as np
```

```python
d1=pd.read_csv('https://raw.githubusercontent.com/nagamounika5/Datasets/master/Advertisement.csv')
d1
```

```python
d1.head()
```

```python
from sklearn.preprocessing import RobustScaler
```

```python
m=RobustScaler()
m
```

```python
m2=m.fit_transform(d1)
m2
```

```python
m2=pd.DataFrame(m2,columns=d1.columns)
m2
```

## standard scaler
 - most commonly used scaling technique.it scales between 0 to mean
## formula
## xi-mean(x)/std(x)
